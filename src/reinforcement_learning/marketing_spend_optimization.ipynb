```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def generate_conversions():
    # Define range of spend values for the x-axis
    # Adapt this range based on your desired curve extent
    spend_values = np.linspace(0, 100, 100)

    # Define a concave function with increasing slope and asymptote at 0
    def cost_function(spend):
        return 100 * np.sqrt(spend) + 10  # Adjust the coefficient and constant for desired slope and asymptote

    # Generate conversions based on the cost function
    conversions = cost_function(spend_values)
    return conversions

# Generate conversions
conversions = generate_conversions()

# Plot using Seaborn
sns.lineplot(x=np.linspace(0, 100, 100), y=conversions)
plt.xlabel('Spend')
plt.ylabel('Conversions')
plt.title('Marketing Spend Optimization')
plt.show()
```
